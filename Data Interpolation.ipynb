```python
# Data Interpolation --> The Process of estimating the new data between the given range of data
# This can be done by three ways---
  # --> Linear Interpolation
  # --> Cubic Interpolation
  # --> Polynomial Interpolation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear Interpolation
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
x_new=np.linspace(1,5,10)
y_inter=np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_inter)
```

```python
# Cubic Interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f=interp1d(x,y,kind='cubic')
x_new=np.linspace(1,5,10)
y_inter=f(x_new)
plt.scatter(x_new,y_inter)
```

```python
# Polynomial Interpolation
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
plt.scatter(x,y)
```

```python
# Interpolate using polynomial interpolation
p=np.polyfit(x,y,2)
x_new=np.linspace(1,5,10)
y_inter=np.polyval(p,x_new)
plt.scatter(x_new,y_inter)
```
