```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Feature Scaling --> It brings the data in one scale
# Standardization >> mu=0 and sigma=1
# Normalization >> (minmax scaling) >> data comes between 0 and 1
# Unit Vector
```

```python
df=sns.load_dataset("tips")
df
```

```python
# Standardization through manually 
Mean=np.mean(df["total_bill"])
std=np.std(df["total_bill"])
std_data=[]
for i in list(df["total_bill"]):
    zscore=(i-Mean)/std
    std_data.append(zscore)
print(std_data)
```

```python
# Feature Scaling affects the distribution of data
```

```python
# sns.distplot(df["total_bill"])
sns.displot(df['total_bill'],kde=True)
```

```python
sns.displot(std_data,kde=True)
```

```python
# Standardization through Library
# mu,std --> .fit() and (i-mu)/std --> .transform()

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler
```

```python
scaler.fit(df[["total_bill"]]) # it will calculate mu and sigma
```

```python
scaler.transform(df[["total_bill"]]) # It throws scaled Data
```
